Sample Use case of freud

```python
#%%
import freud

box, points = freud.data.make_random_system(10, 100, seed=0)
ql = freud.order.Steinhardt(l=6)
ql.compute((box, points), {'r_max':3})

# plot the ql values
import matplotlib.pyplot as plt
plt.plot(ql.particle_order)
plt.show()
```
